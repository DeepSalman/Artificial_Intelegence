Problem 2 - Word Frequency Counter
Write a function word_frequency(text) that takes a string of text and returns a dictionary mapping each word to how many times it appears.
Requirements:
Return the dictionary sorted by frequency, highest first (you can return a list of tuples for this, or a regular dict).
Then write a function most_common_word(text) that reuses word_frequency to return the single most frequent word.



```python
def word_frequency(text):
    words = text.lower().split()
    frequency = {}

    # Count words
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    # Function used for sorting
    def get_frequency(item):
        return item[1]

    # Sort by frequency (highest first)
    sorted_frequency = dict(
        sorted(frequency.items(), key=get_frequency, reverse=True)
    )

    return sorted_frequency
```

```python
text = "apple banana apple orange banana apple"

print(word_frequency(text))
```

Problem 3: Grade Book
Description
You're tracking grades for a class of students. You're given a dictionary where keys are student names and values are lists of their test scores:
grades = { "Alice": [90, 85, 92], 
"Bob": [70, 65, 80], 
"Charlie": [88, 91, 95], 
}

Write these functions:
average(scores): takes a list of numbers, returns the average (float).
student_average(grades, name): returns the average score for one student.
Print a report showing each student's average and letter grade, then print the top student and the overall class average.



```python
grades = {
    "Alice": [90, 85, 92],
    "Bob": [70, 65, 80],
    "Charlie": [88, 91, 95]
}


def average(scores):
    return sum(scores) / len(scores)


def student_average(grades, name):
    return average(grades[name])


def letter_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"


# Print report
print("Student Report")
print("----------------")

top_student = ""
highest_average = 0
total = 0

for student in grades:
    avg = student_average(grades, student)
    grade = letter_grade(avg)

    print(f"{student}: Average = {avg:.2f}, Grade = {grade}")

    total += avg

    if avg > highest_average:
        highest_average = avg
        top_student = student

# Overall class average
class_average = total / len(grades)

print("\nTop Student:", top_student, f"({highest_average:.2f})")
print("Overall Class Average:", f"{class_average:.2f}")
```
