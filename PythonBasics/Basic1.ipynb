Problem 1 - OOP
Description
You're building a simple system to manage a small library. Implement two classes:
Book
Attributes: title, author, isbn, and is_checked_out (defaults to False)
Method checkout(): marks the book as checked out. If it's already checked out, print a message saying so instead.
Method return_book(): marks the book as available again.
Method __str__(): returns a readable string like "1984 by George Orwell (Available)" or "...(Checked Out)".
Library
Attribute: books — a list that holds Book objects (starts empty)
Method add_book(book): adds a Book to the library.
Method find_book(title): searches by title (case-insensitive) and returns the matching Book, or None if not found.
Method checkout_book(title): finds the book by title and checks it out (handle the case where it doesn't exist).
Method list_available_books(): prints all books currently available.
Sample Usage
library = Library() 
library.add_book(Book("1984", "George Orwell", "001")) 
library.add_book(Book("Brave New World", "Aldous Huxley", "002")) library.add_book(Book("Fahrenheit 451", "Ray Bradbury", "003")) 
library.checkout_book("1984") 
library.checkout_book("1984") # should warn it's already checked out library.list_available_books()


```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_checked_out = False

    def checkout(self):
        if self.is_checked_out:
            print(f'"{self.title}" is already checked out.')
        else:
            self.is_checked_out = True
            print(f'You have checked out "{self.title}".')

    def return_book(self):
        self.is_checked_out = False
        print(f'"{self.title}" has been returned.')

    def __str__(self):
        status = "Checked Out" if self.is_checked_out else "Available"
        return f"{self.title} by {self.author} ({status})"
    
class Library:
        def __init__(self):
            self.books = []

        def add_book(self, book):
            self.books.append(book)

        def find_book(self, title):
            for book in self.books:
                if book.title.lower() == title.lower():
                    return book
            return None

        def checkout_book(self, title):
            book = self.find_book(title)
            if book is None:
                print("Book not found.")
            else:
                book.checkout()

        def list_available_books(self):
            print("Available Books:")
            found = False
            for book in self.books:
                if not book.is_checked_out:
                    print(book)
                    found = True

            if not found:
                print("No books are currently available.")
    

```

```python
library = Library()
book1 = Book("1984", "George Orwell", "111")
book2 = Book("The Hobbit", "J.R.R. Tolkien", "222")
book3 = Book("Python Crash Course", "Eric Matthes", "333")
```

```python
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
```

```python
library.list_available_books()
```
