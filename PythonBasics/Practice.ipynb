Problem 1: The Fairest Grade (Lists)
Mr. Rahman is grading a quiz, but he has a rule: he always drops the single highest score before
calculating the class average, so one lucky genius doesn't skew the curve.
Given a list of scores, write code that:
1. Finds the maximum score in the list.
2. Removes one occurrence of that maximum score.
3. Calculates and prints the average of the remaining scores

```python
total_marks = int(input("Enter Number of marks: "))
```

```python
list_of_marks = []
for i in range(total_marks):
    inp = int(input(f"Your {i} input is :"))
    list_of_marks.append(inp)

```

```python

# Use len() to safely get the current list size
n = len(list_of_marks)

for i in range(n):
    # n - i - 1 prevents checking already sorted numbers at the end
    for j in range(0, n - i - 1):
        # Use '>' for ascending order (pushes larger numbers to the right)
        if list_of_marks[j] > list_of_marks[j + 1]:
            # This swaps the two neighboring numbers
            list_of_marks[j], list_of_marks[j + 1] = list_of_marks[j + 1], list_of_marks[j]

list_of_marks
```

```python
#list except the top most 
new_list=list_of_marks[:-1]
new_list
avg=sum(new_list)/(total_marks-1)
avg
```
